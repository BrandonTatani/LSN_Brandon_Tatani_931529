### 32 cities randomly placed on a circumference

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

n_city = 32;

#loading data
x_s, y_s = np.loadtxt('10.1/city.config', usecols = (0,1), delimiter = ' ', unpack = 'true')
x_g, y_g = np.loadtxt('../Esercizi9/9.1/city.config', usecols = (0,1), delimiter = ' ', unpack = 'true')

#plotting data
plt.figure(figsize=(5,5))
plt.scatter(x_s, y_s, label='Simulated Annealing')
plt.scatter(x_g, y_g, label='Genetic Algorithm')
plt.title("Cities' placement")
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
#loading data
T, param = np.loadtxt('10.1/param.dat', usecols=(0,1), delimiter=' ', unpack ='true')
best = np.loadtxt('../Esercizi9/9.1/Best.dat')
a = param[48000-1]
b = best[200-1]
print("Path lenght generated with Simulated Annealing : ", b, "\nPath lenght generated with Genetic Algorithm : ", a)

#plotting data
plt.figure(figsize=(10,5))
plt.subplots_adjust()

plt.subplot(121)
plt.plot(param, label = ('SA Generated'))
plt.title('Best path over time')
plt.xlabel('Step')
plt.ylabel('Path lenght')
plt.legend()

plt.subplot(122)
plt.plot(best, label = 'GA Generated')
plt.title('Best path over time')
plt.xlabel('Generation')
plt.ylabel('Path lenght')
plt.legend()

plt.show()
```

The two algorithms have produced the same results, in a comperable computational time.
