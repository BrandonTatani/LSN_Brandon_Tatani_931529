```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

M=100000
N=100

x = np.arange(N)
av = np.zeros(N)
err = np.zeros(N)
tmp = 0
i=0
file = open('output.dat', 'r')

for line in file:
    columns = line.split()
    x[i]= columns[0]
    av[i] = columns[1]
    err[i] = columns[2]
    i+=1
    
plt.errorbar(x, av, err)
plt.grid(True)
plt.show()
```
