### 32 cities randomly placed on a circumference

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

n_city = 32;

#loading data
x, y = np.loadtxt('9.1/city.config', usecols = (0,1), delimiter = ' ', unpack = 'true')

#plotting data
plt.figure(figsize=(5,5))
plt.scatter(x, y)
plt.title("Cities' placement")
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
n_gen = 200
raggio = 10

#loading data
n = np.arange(n_gen)
cir = np.ones(n_gen)*2*3.14*raggio
best = np.loadtxt('9.1/Best.dat')
half =np.loadtxt('9.1/half_ave.dat')

#plotting data
plt.figure(figsize=(7,7))

plt.plot(best, label = 'Best path')
plt.plot(half, label = 'Average on best half population')
plt.plot(cir , label =('Lenght of the circumference'))
plt.title('Best path over time')
plt.xlabel('Generation')
plt.ylabel('Path lenght')
plt.legend()

plt.show()
```
