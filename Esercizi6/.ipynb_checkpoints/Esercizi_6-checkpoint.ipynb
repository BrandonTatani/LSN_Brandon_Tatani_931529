### Exercise 06.1

- <span style="color:red">Show pictures of your estimations of $U(N,T)$, $C(N,T)$, $\chi(N,T)$, $M(N,T,h=0.02)$ and their uncertainties</span> obtained with a large number of *MC steps* (e.g. $M\ge 10^5$) compared with the exact curves. In the same picture show your results obtained with both sampling methods, Metropolis and Gibbs.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

#variabili

points=100
T = np.linspace(0.2,3.0,num=points)
beta = 1/T
J = 1.0
Ns = 50

#loading data

t, metro, err_m = np.loadtxt("metro/ene.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
gibbs, err_g = np.loadtxt("gibbs/ene.dat", usecols=(1,2), delimiter=' ', unpack='true')

th = np.tanh(J/T)
thN= th**Ns
ch = 1/th
e = -J*( th + ch*thN )/( 1 + thN )

#plotting data
plt.figure(figsize=(10,5))
plt.subplots_adjust()

plt.subplot(121)
plt.errorbar(t, metro, yerr=err_m, label='Metropolis')
plt.plot(T, e, label='Esatto')
plt.title('Ising 1D, internal energy')
plt.xlabel('T')
plt.ylabel('U/N')
plt.legend()

plt.subplot(122)
plt.errorbar(t, gibbs, yerr=err_g, label='Gibbs')
plt.plot(T, e, label='Esatto')
plt.title('Ising 1D, internal energy')
plt.xlabel('T')
plt.ylabel('U/N')
plt.legend()

plt.tight_layout()
plt.show();
```

```python
#Heat Capacity

#loading data
heat=((beta*J)**2)*(((1+thN+(Ns-1)*(th**2)+(Ns-1)*(ch**2)*thN)/(1+thN))-Ns*((th+ch*thN)/(1+thN))**2)
t, metro, err_m = np.loadtxt("metro/heat.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
gibbs, err_g = np.loadtxt("gibbs/heat.dat", usecols=(1,2), delimiter=' ', unpack='true')

#plotting data
plt.figure(figsize=(10,5))
plt.subplots_adjust()

plt.subplot(121)
plt.errorbar(t, metro, yerr=err_m, label='Metropolis')
plt.plot(T, heat, label='esatto')
plt.title('Ising 1D, Heat capacity')
plt.xlabel('T')
plt.ylabel('C')
plt.legend()

plt.subplot(122)
plt.errorbar(t, gibbs, yerr=err_g, label='Gibbs')
plt.plot(T, heat, label='esatto')
plt.title('Ising 1D, Heat capacity')
plt.xlabel('T')
plt.ylabel('C')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
#Magnettizzazione

#variabili
h=0.02 #external field
b = 1/T

#loading data
l1 = np.exp(b*J)*np.cosh(b*h)+np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
l2 = np.exp(b*J)*np.cosh(b*h)-np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
Z = l1**Ns + l2**Ns
M = (np.exp(b*J)*np.sinh(b*h)*((l1**(Ns-1))*(1+np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))) 
        + (l2**(Ns-1))*(1-np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J)))))/(Z)

t, metro, err_m = np.loadtxt("metro/mag.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
gibbs, err_g = np.loadtxt("gibbs/mag.dat", usecols=(1,2), delimiter=' ', unpack='true')
metro_2, err_m2 = np.loadtxt("metro/mag_2.dat", usecols=(1,2), delimiter=' ', unpack='true')
gibbs_2, err_g2 = np.loadtxt("gibbs/mag_2.dat", usecols=(1,2), delimiter=' ', unpack='true')
y = np.zeros(16)

#plotting data
plt.figure(figsize=(10,10))
plt.subplots_adjust()

plt.subplot(221)
plt.errorbar(t, metro, yerr=err_m, label='Metropolis')
plt.plot(t, y, label='esatto')
plt.title('Ising 1D, magnetization M with h = 0.0')
plt.xlabel('T')
plt.ylabel('$M$')
plt.legend()

plt.subplot(222)
plt.errorbar(t, gibbs, yerr=err_g, label='Gibbs')
plt.plot(t, y, label='esatto')
plt.title('Ising 1D, magnetization M with h = 0.0')
plt.xlabel('T')
plt.ylabel('$M$')
plt.legend()

plt.subplot(223)
plt.errorbar(t, metro_2, yerr=err_m2, label='Metropolis')
plt.plot(T, M, label='esatto')
plt.title('Ising 1D, magnetization M with h = 0.2')
plt.xlabel('T')
plt.ylabel('$M$')
plt.legend()

plt.subplot(224)
plt.errorbar(t, gibbs_2, yerr=err_g2, label='Gibbs')
plt.plot(T, M, label='esatto')
plt.title('Ising 1D, magnetization M with h = 0.2')
plt.xlabel('T')
plt.ylabel('$M$')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
#Suscettibilit magentica

#loading data
X = beta*np.exp(2*beta*J)*(1-thN)/(1+thN)
t, metro, err_m = np.loadtxt("metro/chi.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
gibbs, err_g = np.loadtxt("gibbs/chi.dat", usecols=(1,2), delimiter=' ', unpack='true')

#plotting data
plt.figure(figsize=(10,5))
plt.subplots_adjust()

plt.subplot(121)
plt.errorbar(t, metro, yerr=err_m, label='Metropolis')
plt.plot(T, X, label='esatto')
plt.title('Ising 1D, Susceptibility')
plt.xlabel('T')
plt.ylabel('$\chi$')
plt.legend()

plt.subplot(122)
plt.errorbar(t, gibbs, yerr=err_g, label='Gibbs')
plt.plot(T, X, label='esatto')
plt.title('Ising 1D, Susceptibility')
plt.xlabel('T')
plt.ylabel('$\chi$')
plt.legend()

plt.tight_layout()
plt.show()
```
